```python
from ISLP import load_data
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
auto = load_data("auto")
auto
```

```python
auto = auto.dropna()
auto
```

# a) 

```python
auto.dtypes
```

```python
quant = auto.select_dtypes(include=['float64', 'int64'])
quant
```

```python
qual = auto.select_dtypes(include=['object'])
qual
```

# b)

```python
col = ['mpg','cylinders','displacement','horsepower','weight','acceleration','year']
min = auto[col].min()
max = auto[col].max()
print(f"Range of {col}: {min} , {max}")
```

# C)

```python
mean = auto[col].mean()
std = auto[col].std()
print(f"{col}: {mean} , Std Dev of {col}: {std}")
```

# d) 

```python
auto_subset = auto.drop(auto.index[9:85])
```

```python
min = auto_subset[col].min()
max = auto_subset[col].max()
mean = auto_subset[col].mean()
std = auto_subset[col].std()
print(f"Range of {col}: {min} , {max}")
print(f"Mean of {col}: {mean} , Std Dev of {col}: {std}")
```

# e)

```python
sns.pairplot(auto[['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'year']])
plt.suptitle("Pairplot of Quant Predictors", y=1.02)
plt.show()
```

# f)

```python
sns.scatterplot(x='horsepower', y='mpg', data=auto)
plt.title('Horsepower vs mpg')
```

```python
sns.scatterplot(x='weight', y='mpg', data=auto)
plt.title("mpg vs Weight")
```

```python
sns.scatterplot(x='year', y='mpg', data=auto)
plt.title("mpg vs Year")
```

```python
sns.boxplot(x='cylinders', y='mpg', data=auto)
plt.title("Cyl")
```

Yes, some variables help predict mpg.

Horsepower, weight, and displacement go up, mpg goes down.

Year goes up, mpg goes up (newer cars are better).

Fewer cylinders = better mileage.
